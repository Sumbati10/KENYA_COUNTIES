```python
import requests
import csv

# Base URL for querying the ArcGIS feature service
base_url = "https://services6.arcgis.com/zOnyumh63cMmLBBH/ArcGIS/rest/services/Kenya_Soils_Data/FeatureServer/"
layer_id = "0"  # Specify the layer ID to query

# Construct the query URL to fetch all features in GeoJSON format
query_url = f"{base_url}{layer_id}/query?where=1%3D1&outFields=*&outSR=4326&f=pgeojson"

# Make a GET request to the ArcGIS feature service
response = requests.get(query_url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print("Data received successfully.")
    
    # Get properties from the first feature (for demonstration)
    properties = data['features'][0]['properties']
    
    # Write the properties to a CSV file
    with open('kenya_soils_DATA5.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        
        # Write headers (keys of the properties dictionary)
        writer.writerow(list(properties.keys()))
        
        # Write the first row of data (values of the properties dictionary)
        writer.writerow(list(properties.values()))
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
import requests

url = "https://services6.arcgis.com/zOnyumh63cMmLBBH/ArcGIS/rest/services/Kenya_Soils_Data/FeatureServer/0/query"
params = {
    "where": "1=1",
    "outFields": "*",
    "f": "geojson"
}

try:
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        with open("kenya_soils_data10.geojson", "w") as file:
            file.write(response.text)
        print("GeoJSON data successfully saved.")
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
except Exception as e:
    print(f"An error occurred: {e}")

```
